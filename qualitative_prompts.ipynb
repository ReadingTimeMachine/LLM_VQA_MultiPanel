From https://arxiv.org/html/2509.04457v1
* Models are explicitly prompted to articulate a logical pathway to a known ground-truth answer. The prompt requires the model to structure its output with specific XML-style tags, separating the reasoning traces (<thought>) from the final output (<answer>), while strictly forbidding the model from mentioning the "original answer" inside the thought process to prevent cheating
