```python
# get model names
model_names = []
for k,v in replace_names.items():
    if k+'/' in dirs_to_use:
        model_names.append(v)

model_names
```

```python
# loop and count NaNs
count_nans = {}
for m in model_names:
    count_nans[m] = 0
for model in model_names:
    d = df[df['model']==model]
    # print('model =', model)
    for i,dd in d.iterrows():
        for k,v in dd['LMM Answer'].items():
            if not isinstance(v, list):
                if pd.isnull(v) or v is None:
                    count_nans[model] += 1
            else:
                for vv in v:
                    if not isinstance(vv,list):
                        if pd.isnull(vv) or vv is None:
                            count_nans[model] += 1
                    else:
                        for vvv in vv:
                            if not isinstance(vvv,list):
                                if pd.isnull(vvv) or vvv is None:
                                    count_nans[model] += 1
                            else:
                                for v2 in vvv:
                                    if pd.isnull(v2) or v2 is None:
                                        count_nans[model] += 1

count_nans
```
