```python
# get model names
model_names = []
for k,v in replace_names.items():
    if k+'/' in dirs_to_use:
        model_names.append(v)

model_names
```

```python
# loop and count NaNs
count_nans = {}
for m in model_names:
    count_nans[m] = 0
for model in model_names:
    d = df[df['model']==model]
    # print('model =', model)
    for i,dd in d.iterrows():
        for k,v in dd['LMM Answer'].items():
            if not isinstance(v, list):
                if pd.isnull(v) or v is None:
                    count_nans[model] += 1
            else:
                for vv in v:
                    if not isinstance(vv,list):
                        if pd.isnull(vv) or vv is None:
                            count_nans[model] += 1
                    else:
                        for vvv in vv:
                            if not isinstance(vvv,list):
                                if pd.isnull(vvv) or vvv is None:
                                    count_nans[model] += 1
                            else:
                                for v2 in vvv:
                                    if pd.isnull(v2) or v2 is None:
                                        count_nans[model] += 1

count_nans
```

```python
# now read in saved tokens
# cost
cost_input = 1.00 # per million
cost_output = 5.00 # per million

claude_tokens = {}
for m in model_names:
    if 'claude' in m.lower():
        claude_tokens[m] = {'input':0, 'output':0}
for d in dirs_to_use:
    print('')
    print('****************************')
    print(d)
    files = glob(main_llm_dir + d + '*pickle')
    for ifile,file in enumerate(files):
        with open(file, 'rb') as f:
            data = pickle.load(f)

        # if ifile == 0: 
        #     print('n elements:', len(data))
        #     # print(data[0])
        if 'claude' in d.lower():
            model = replace_names[d.replace('/','')]
            questions = data[0]
            for q in questions:
                claude_tokens[model]['input'] += q['usage'].input_tokens
                claude_tokens[model]['output'] += q['usage'].output_tokens
            # import sys; sys.exit()
```
