```python
table_dir = '~/LLM_VQA_MultiPanel/figure_notebooks/'

table_file = 'other_datasets_chart_asLatex.txt'
```

```python
from astropy.table import Table
import os

table_dir = os.path.expanduser(table_dir)
```

```python

# Read a .tex file or a string containing the table
dat = Table.read(table_dir+table_file, format='ascii.latex')

# Convert to Pandas DataFrame
df = dat.to_pandas()
```

```python
# new table based on if table
chart_types = []
for i,r in df.iterrows():
    ct = ''
    for cn,tag in zip(['Line', 'Bar', 'Sctr', 'Pie', 'Hist', 'Othr'], ['l','b','s','p','h','o']):
        if 'checkmark' in r[cn]:
            ct += tag +','
    ct = ct.removesuffix(',')
    chart_types.append(ct)
```

```python
df_table = df.copy()[['Name', 'Date', 'BB', 'Data Link', 'MultiPanel', 'Citation(s)']]
df_table['Plot Type'] = chart_types
# reorder
df_table = df_table[['Name', 'Date', 'Plot Type', 'BB', 'Data Link', 'MultiPanel', 'Citation(s)']]
```
