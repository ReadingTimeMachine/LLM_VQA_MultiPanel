This is to get URLs and arxiv links for papers in a .bib file.

To export collection, right-click on collection and hit "Export Collection..."

```python
collection_dir = '~/Downloads/'
# collection_files = ['Background-LLM:VQA:IR in Astronomy & Science.bib'] 
collection_files = ['iConfSimilar.bib'] 
```

```python
import os
from glob import glob
```

```python
collection_dir = os.path.expanduser(collection_dir)
```

```python
urls = []

for cf in collection_files:
    filepath = collection_dir + cf
    with open(filepath, 'r') as f:
        text = f.readlines()

    # loop lines
    for l in text: 
        if 'url={' in l.replace(' ', ''):
            #print(l)
            ll = l.split('url = {')[-1].split('}')[0]
            print(ll)
            urls.append(ll)
```
