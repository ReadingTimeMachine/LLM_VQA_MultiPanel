```python
collection_dir = '~/Downloads/'
#collection_files = ['Background.bib', 'SyntheticFigures.bib', 'JCDL 2024 Figures with LLMs.bib', 'Background2.bib'] # if only .bib files are exported
collection_files = ['Background-LLM:VQA:IR in Astronomy & Science.bib']
#collection_dirs = ['synthfigs/'] # if there is a directory
collection_dirs = [] # if there is a directory

# where to copy to?
notebooklm_dir = '~/Downloads/tmp/toNLM2/'
# for saving to google drive
#notebooklm_dir = '~/Library/CloudStorage/GoogleDrive-jnaiman@illinois.edu/My Drive/NASA_ADAP_2026/literature/'
```

```python
import os
from shutil import copyfile
from glob import glob
```

```python
collection_dir = os.path.expanduser(collection_dir)
notebooklm_dir = os.path.expanduser(notebooklm_dir)
```

```python
if not os.path.exists(notebooklm_dir):
    print("making:", notebooklm_dir)
    os.mkdir(notebooklm_dir)
```

```python
for cf in collection_files:
    filepath = collection_dir + cf
    with open(filepath, 'r') as f:
        text = f.readlines()
    # loop lines
    for l in text: 
        if 'file=' in l.replace(' ', ''):
            #print(l)
            ll = l.split('file = {')[-1].split('}')[0]
            # take out any snapshots
            ll = ll.split(':text/html;')[-1]
            ll = ll.split(':')[1]
            llout = ll.split('/')[-1]


            # check html
            if '.html' in ll:
                import sys; sys.exit()
            if not os.path.exists(notebooklm_dir + llout):
                try:
                    copyfile(ll,notebooklm_dir + llout)
                except:
                    print('could not copy:', ll)


            
    
# for cfd in collection_dirs:
#     filepath = collection_dir + cfd
#     files = glob(filepath + 'files/*/*')
#     for f in files:
#         if not os.path.exists(notebooklm_dir)
#     print(files)
```
